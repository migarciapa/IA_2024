```python
## RECONOCIMIENTO DE VOZ LOCAL DE COMANDOS
## David Miguel Garcia Palacios
## Universidad Nacional de Colombia Sede Bogota

# Instalacion e importacion de bibliotecas
!pip install vosk
!pip install pydub
import IPython.display as ipd
import numpy as np
import vosk
import pydub
import os

# Descarga y definicion del modelo de vosk
if not os.path.isdir("/content/vosk-model/vosk-model-small-es-0.42"):
  !wget https://alphacephei.com/vosk/models/vosk-model-small-es-0.42.zip
  !unzip vosk-model-small-es-0.42.zip -d vosk-model
  os.remove("/content/vosk-model-small-es-0.42.zip")
vosk_model = vosk.Model("/content/vosk-model/vosk-model-small-es-0.42")

# Obtencion del audio
sample_rate = 16000
audio = pydub.AudioSegment.from_file("/content/muestra.wav")
audio = audio.set_channels(1).set_frame_rate(sample_rate)
audio = np.array(audio.get_array_of_samples())
audio_data = audio.tobytes()

# Reconocimiento de voz
recognizer = vosk.KaldiRecognizer(vosk_model,sample_rate)
recognizer.AcceptWaveform(audio_data)
comando = recognizer.Result()

# Obtencion del resultado
print(comando)
ipd.Audio(data=audio,rate=sample_rate)
```
